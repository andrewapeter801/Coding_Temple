1. Which hosts are busiest and why?

```python
df = pd.read_csv(r'C:\Users\bioni\OneDrive\Documents\GitHub\Coding_Temple\Week 3\Assignments\Wkend Project\AB_NYC_2019.csv')

df.columns
```

```python
df.groupby(['host_name']).count().sort_values('availability_365', ascending = False)[['calculated_host_listings_count','availability_365','reviews_per_month']]
```

The hosts with the most properties and highest reviews are typically the busiest, they are also among the most with no availability. john was the busiest because he had the most amount of properties and reviews per month.


```python
# How many neighborhood groups are available and which shows up the most?

df.groupby(['neighbourhood_group']).count().sort_values(['neighbourhood'], ascending=False)
```

Manhatten and Brooklyn have, by far, the most listings by count. Manhatten tops the charts with the most listings overall.

```python
# Are private rooms the most popular in manhattan?

df.groupby(['neighbourhood_group','room_type',]).count().sort_values('availability_365', ascending= False)[['calculated_host_listings_count']]
```

Based on our review, entire home/apt are the most popular in Manhatten with 13,199 listings.
Private rooms were second in popularity with 7,982.

```python
# Which hosts are the busiest and based on their reviews?

df.groupby(['host_name']).count().sort_values('reviews_per_month', ascending = False)[['calculated_host_listings_count','availability_365','reviews_per_month']]
```

Based on our review; David, Michael, Alex, and Sonder were the hosts that were the busiest based on thier calculated host rating and reviews per month. Overall, John had the highest occupancy rate per property, per review each month.

```python
#Which neighorhood group has the highest average price?
df.groupby(['neighbourhood_group']).mean(True).sort_values(['price'],ascending = False)[['price']].round(2)
```

Manhatten had the highest average price at $196.88 per night.

```python
# Which neighbor hood group has the highest total price?
df.groupby(['neighbourhood_group']).max(True).sort_values(['price'],ascending = False)[['price']]
```

Manhatten had the highest total price for all listings combined. However, it is interesting to note that; Manhatten, Queens, and Brooklyn all had listings that entertained $10,000 per night listings. 

```python
#Which top 5 hosts have the highest total price?
df.groupby(['host_name']).max(True).sort_values(['price'],ascending = False)[['price']].head(5)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
conditional = (df['number_of_reviews'] >= 100) & (df['availability_365']==0)
df[conditional].groupby(['host_name']).count().sort_values(['number_of_reviews'], ascending=False)[['number_of_reviews','availability_365']]

```

141 had zero availability with a review count of 100 or more. 

```python
# What host has the highest total of prices and where are they located?

df.groupby(['neighbourhood_group','neighbourhood','host_name']).max(numeric_only=True).sort_values(['price'], ascending = False)[['price']].head(1)
```

```python
# When did Danielle from Queens last receive a review?

df[df['host_name']=='Danielle'].groupby(['last_review']).max(False).sort_values(['number_of_reviews'],ascending = False).head(5)

```
