## Correlations

```python
# Get correlations for numeric data
auto.corr(numeric_only= True)
```

```python
# Create a heatmap
# Establish size of figure.
plt.figure(figsize= (16,9))
# Get correlation of variables.
corr = auto.corr(numeric_only= True)
# Set up mask to be "True" in the upper triangle.
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

# Plot our correlation heatmap
sns.heatmap(corr,
            square= True,
            cmap= 'coolwarm',
            annot= True,
            vmax=1,
            vmin=-1,
            mask=mask);
```

```python
# What if we ONLY wanted to look at how the features correlate with `mpg`?
#auto.corr(numeric_only= True)[['mpg']].sort_values(by='mpg', ascending=False)
sns.heatmap(auto.corr(numeric_only= True)[['mpg']].sort_values(by='mpg', ascending=False),
            annot = True,
            vmax=1,
            vmin=-1,
            cmap='coolwarm');
```
